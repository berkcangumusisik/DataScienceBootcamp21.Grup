## Matplotlib Kütüphanesi
 
Matplotlib, Python programlama dilinde veri görselleştirme için en yaygın kullanılan kütüphanelerden biridir. Özellikle bilimsel ve mühendislik alanlarında verilerin grafiklerle görselleştirilmesi amacıyla geliştirilmiştir. Temel olarak 2D (iki boyutlu) grafikler çizmek için kullanılır, fakat bazı ek kütüphaneler sayesinde 3D grafikler de oluşturulabilir.

### Temel Özellikleri

- **Çeşitli Grafik Türleri**: Çizgi grafiği (line plot), çubuk grafiği (bar plot), histogram, pasta grafiği (pie chart), scatter plot (dağılım grafiği) gibi birçok görselleştirme türünü destekler.
- **Yüksek Özelleştirilebilirlik**: Grafiklerin başlıkları, eksen etiketleri, renkleri, işaretleyicileri (marker), çizgi tipi ve kalınlığı gibi birçok özelliği kolayca değiştirilebilir, özelleştirilebilir.
- **Dışa Aktarma**: Hazırlanan grafikler yüksek çözünürlükte PNG, PDF, SVG gibi farklı dosya formatlarında kaydedilebilir.
- **Diğer Kütüphaneler ile Entegrasyon**: NumPy ve Pandas gibi veri analiz kütüphaneleriyle sorunsuz çalışır.

### Temel Kullanımı

Matplotlib'in en çok kullanılan alt modülü `pyplot`'tur. `pyplot`, Matlab benzeri kolay bir sözdizimi sunar ve çoğunlukla şöyle içe aktarılır:
 
```python
 import matplotlib.pyplot as plt
```
Temel bir çizgi grafiği oluşturmak için örnek kullanım:
# 
```python
import matplotlib.pyplot as plt
 
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.title("Basit Çizgi Grafiği")       # Grafik başlığı
plt.xlabel("X Ekseni")                 # Yatay eksenin adı/etiketi
plt.ylabel("Y Ekseni")                 # Dikey eksenin adı/etiketi
plt.show()                             # Grafiği ekranda göster
```
 
**Açıklamalar:**
- `plt.plot(x, y)`: x ve y veri dizilerini birleştirerek bir çizgi grafiği çizer.
- `plt.title("...")`: Grafik üst kısmındaki başlığı belirler.
- `plt.xlabel("...")`: X ekseni (yatay eksen) için açıklayıcı bir etiket ekler. Özellikle eksenin temsil ettiği verinin ne olduğunu göstermek için kullanılır. Grafik yorumu ve okunabilirliği açısından önemlidir.
- `plt.ylabel("...")`: Y ekseni (dikey eksen) için açıklayıcı bir etiket ekler. Yine, okuyan kişinin hangi veriyi gördüğünü anlamasına yardımcı olur.
- `plt.show()`: Tüm grafik ayarlarından sonra grafiği ekranda görüntüler.

Özetle, eksen isimleri (`xlabel` ve `ylabel`), grafik üzerindeki eksenlerin neyi gösterdiğini açıklamak ve grafiği daha anlaşılır hale getirmek için kullanılır. Eksensiz grafikler, özellikle karmaşık ya da çok veri içeren çalışmalarda yoruma ve anlaşılmaya engel olabilir.

Matplotlib; veri analizi, raporlama ve bilimsel çalışmalar için güçlü, esnek ve kapsamlı bir görselleştirme altyapısı sunar. Kapsamlı dokümantasyonu ve geniş topluluk desteği sayesinde, Python'da veri görselleştirme dendiğinde ilk akla gelen kütüphanelerden biridir.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
df = sns.load_dataset("titanic")
df.head()
````

````python
df["sex"].value_counts().plot(kind="bar")
plt.show()
````

````python
plt.hist(df["age"])
````

````python
plt.boxplot(df["fare"])
plt.show()
````

````python
x = np.array([1,8])
y = np.array([0,150])
plt.plot(x,y)
plt.show()
````

````python
plt.plot(x,y, "o")
plt.show()
````

````python
y = np.array([13,28,11,100])
plt.plot(y, marker = "o")
plt.show()
````

````python
markers = ["o", "v", "^", "<", ">", "s", "p", "*", "h", "H", "D", "d", "P", "X"]
for marker in markers:
    plt.plot(y, marker = marker)
    plt.show()
````

````python
plt.plot(y, linestyle = "dotted") # dotted, dashed, dashdot, solid
plt.show()
````

````python
x = np.array([23, 18, 31, 10])
y = np.array([13, 28, 11, 100])
plt.plot(x)
plt.plot(y)
plt.show()
````

````python
k = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(k,y)
plt.title("Grafik Başlığı")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.grid(True)
plt.show()
````

````python
#######################
# Subplots
#######################

# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 2, 1)
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 2, 2)
plt.title("2")
plt.plot(x, y)
plt.show()
````

````python
# 3 grafiği bir satır 3 sütun olarak konumlamak.
# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 1)
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 3, 2)
plt.title("2")
plt.plot(x, y)

# plot 3
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 3)
plt.title("3")
plt.plot(x, y)

plt.show()
````
