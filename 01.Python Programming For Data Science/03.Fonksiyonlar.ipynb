### Fonksiyon Nedir?

- Belirli bir işi yapan, tekrar tekrar kullanılabilen kod bloklarıdır.
- Bir kez tanımlanır, ihtiyaç duyulduğunda çağrılır.
- Kodun okunabilirliğini artırır ve hatayı azaltır.
- Parantez ile çağrılır; parantez içindeki değerlere **argüman** denir.
- Argümanları alır, işlem yapar ve gerektiğinde sonuç döndürür.

### Fonksiyonlar Ne İşe Yarar?

- Aynı işlemi tekrar yazmak yerine tek yerde tanımlamayı sağlar.
- Kodu küçük ve yönetilebilir parçalara böler.
- Python ile birlikte gelen yerleşik fonksiyonlar doğrudan kullanılabilir.
- `def` ile kendi fonksiyonlarımızı tanımlayabiliriz.
- **Fonksiyon** bağımsız çağrılır; **metot** bir nesneye bağlıdır ve nokta ile çağrılır.

### Fonksiyon Nasıl Tanımlanır?

- `def` anahtar kelimesiyle tanımlanır.
- Sırasıyla: fonksiyon adı, parantez, parametreler, iki nokta üst üste.
- Alt satırdan itibaren girintili olarak gövde yazılır.
- Parantez içindeki değişkenlere **parametre** denir.
- Çağrılırken verilen değerlere **argüman** denir.
- Sonuç döndürmek için `return` kullanılır.

```python
def fonksiyon_adi(parametre1, parametre2):
    # yapılacak işlemler
    return sonuc
```

### Bir Parametreli Fonksiyon Nasıl Tanımlanır?

- Parantez içinde yalnızca **bir** değişken adı yazılır.
- Bu parametre, dışarıdan alınan tek bilgiyi temsil eder.
- Gövdede bu parametre kullanılarak işlem yapılır.
- Çağrılırken parantez içine **tek** argüman verilir.

```python
def fonksiyon_adi(parametre):
    # yapılacak işlemler
    return sonuc
```

### İki Parametreli Fonksiyon Nasıl Tanımlanır?

- Parantez içinde **virgülle ayrılmış iki** değişken adı yazılır.
- Bu parametreler, dışarıdan alınan iki ayrı bilgiyi temsil eder.
- Gövdede her iki parametre birlikte kullanılır.
- Çağrılırken parantez içine **sırasıyla iki** argüman verilir.

```python
def fonksiyon_adi(parametre1, parametre2):
    # yapılacak işlemler
    return sonuc
```

### Özet

- Fonksiyonlar `def` ile tanımlanır.
- **Bir parametreli:** parantezde tek değişken, çağrıda tek argüman.
- **İki parametreli:** parantezde iki değişken, çağrıda iki argüman.
- Parametre sayısı, fonksiyonun kaç bilgiye ihtiyaç duyduğuna göre belirlenir.

`````python
def calculate(x):
    print(x * 2)
calculate(2)
`````

`````python
def summer(arg1,arg2):
    print(arg1+arg2)

summer(1,2)
`````

### Docstring Nedir?

- Docstring, bir fonksiyon, sınıf veya modülün ne yaptığını açıklayan özel bir metin bloğudur.
- Fonksiyonun hemen altında, üç çift tırnak arasında yazılır (`""" açıklama """`).
- Fonksiyonun amacını, parametrelerini ve döndürdüğü değeri belirtmek için kullanılır.
- Python'un yerleşik `help()` fonksiyonu ile fonksiyon hakkında bilgi almak için docstring kullanılır.

Örnek:

```python
def karesini_al(x):
    """
    Verilen x sayısının karesini döndürür.

    Parametreler:
      x (int veya float): Karesi alınacak sayı

    Döndürür:
      int veya float: x'in karesi
    """
    return x ** 2
```

`````python
def summer(arg1,arg2):
    """
    Sum two numbers
    Parameters/Args:
    -------------
    arg1: int, float
    arg2: int, float

    Returns:
    --------
    int, float: arg1 ve arg2'nin toplamı

    """
    print(arg1+arg2)

summer(1,2)



`````

### Ön Tanımlı Argümanlar (Default Arguments)

Fonksiyonlara parametre tanımlarken, bazı parametrelere varsayılan (ön tanımlı) değerler atayabiliriz. Böylece, kullanıcı bu parametreye değer vermezse, otomatik olarak belirlediğimiz değeri kullanır. Eğer kullanıcı bir değer verirse, verilen değer kullanılır.

Aşağıda bir örnek görebilirsiniz:

```python
def say_hello(name="Ziyaretçi"):
    print(f"Merhaba, {name}!")

say_hello()           # Merhaba, Ziyaretçi!
say_hello("Ahmet")    # Merhaba, Ahmet!
```

Burada `name` parametresinin varsayılan değeri `"Ziyaretçi"` olarak tanımlanmıştır. Fonksiyon çağrılırken `name` belirtilmezse bu değer kullanılır.

Varsayılan argümanlar, fonksiyonları daha esnek ve kullanımı kolay hale getirir. Bir fonksiyonda birden fazla varsayılan argüman tanımlanabilir, ancak varsayılan değer verilen parametrelerin, parametre listesinde en sonda olmasına dikkat edilmelidir.

```python
def bilgiler(ad, soyad, yas=18):
    print(f"Ad: {ad}, Soyad: {soyad}, Yaş: {yas}")

bilgiler("Ayşe", "Yılmaz")         # Yaş için varsayılan değer kullanılır (18)
bilgiler("Mehmet", "Demir", 25)   # Yaş için verilen değer kullanılır (25)
```

`````python
def divide(a,b = 1):
    print(a/b)

divide(10)
`````

- Birden çok yerde değiştirilebilmesi kolay bir yapıya ihtiyaç duyduğumuzda fonksiyonları kullanırız.

### return Nedir? Fonksiyonlarda Kullanımı

`return` ifadesi, bir fonksiyonda sonucu fonksiyonun dışına döndürmek için kullanılır. Yani, fonksiyon çalışmasını bitirdiğinde `return` ile belirli bir değeri (veya değerleri) çağrılan yere iletebilir.

`print` fonksiyonun çıktısını ekrana yazdırırken, `return` değeri geri döndürür ve kodun başka bir yerinde kullanılmasını sağlar.

Fonksiyonlarda `return` ifadesinin nasıl kullanıldığına bir örnek:

```python
def toplama(a, b):
    return a + b

sonuc = toplama(3, 5)
print(sonuc)   # 8
```

Yukarıda, `toplama` fonksiyonu iki sayının toplamını döndürür. `return` anahtar kelimesi ile dönen değeri, fonksiyon çağrılırken kullanabilir veya bir değişkene atayabiliriz.

Bir fonksiyonda `return` ifadesi çalıştığında, fonksiyonun kalan kısmı çalıştırılmaz ve fonksiyondan çıkılır.

Birden fazla değeri aynı anda döndürmek için ise `return` ile birden fazla değer virgülle ayrılarak yazılabilir. Bu durumda dönen değerler bir demet (`tuple`) olur:

```python
def islemler(a, b):
    toplam = a + b
    carpim = a * b
    return toplam, carpim

t, c = islemler(4, 5)
print(t)  # 9
print(c)  # 20
```

`````python
def calculate(varm, moisture, charge):
    return varm * moisture * charge

a = calculate(10, 20, 30)
print(a)

`````

## Fonksiyon İçinde Fonksiyon Çağırmak

`````python
def calculate(varm, moisture, charge):
    return int((varm + moisture) / charge)

calculate(10, 20, 30) * 10
`````

`````python
def standardize(a,p):
    return a * 10 / 100 * p * p
`````

`````python
def all_calculate(varm, moisture, charge,p):
    a = calculate(varm, moisture, charge)
    b = standardize(a,p)
    print(b * 10)

all_calculate(1,3,5,12)



`````

### Yerel (Local) ve Global Değişkenler

Bir fonksiyon içinde tanımlanan değişkenlere **yerel (local) değişken** denir. Bu değişkenlerin geçerlilik alanı sadece ilgili fonksiyon içindedir; fonksiyon dışında çağrılamazlar. Fonksiyon dışında tanımlanan değişkenlere ise **global değişken** denir ve bu değişkenlere tüm kod boyunca erişilebilir.

Örnek:
```python
x = 10  # Global değişken

def fonksiyon():
    y = 5  # Yerel değişken
    print("Fonksiyon içinden x:", x)  # Global değişkene erişim mümkün
    print("Fonksiyon içinden y:", y)

fonksiyon()
print("Fonksiyon dışından x:", x)  # Global değişkene erişim mümkün
# print("Fonksiyon dışından y:", y)  # Hata: y tanımlı değil
```

Fonksiyon içinde global bir değişkenin değerini değiştirmek isterseniz `global` anahtar kelimesini kullanabilirsiniz:

```python
x = 10

def fonksiyon():
    global x
    x = 20

fonksiyon()
print(x)  # Sonuç: 20
```

Kısacası:
- Global değişkenler fonksiyonun dışında tanımlanır, her yerden erişilebilir.
- Yerel değişkenler fonksiyonun içinde tanımlanır, sadece fonksiyon içinde geçerlidir. Dışarıdan erişilemezler.

`````python
list_scores = [1,2]

def add_element(a,b):
    c = a * b
    list_scores.append(c)
    print(list_scores)

add_element(10,20)
`````
