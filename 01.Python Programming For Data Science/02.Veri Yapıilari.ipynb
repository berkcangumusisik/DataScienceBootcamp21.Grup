## Sayılar (Numbers)

Python'da **sayılar** (numbers) temel veri tiplerinden biridir ve sayısal işlemler için sıklıkla kullanılır. En yaygın kullanılan sayı tipleri şunlardır:

- **int**: Tam sayılar (örnek: `5`, `-3`, `100`)
- **float**: Ondalıklı sayılar (örnek: `3.14`, `-7.5`, `0.0`)
- **complex**: Karmaşık sayılar (örnek: `2 + 3j`)

### Temel Sayısal İşlemler

Python, aritmetik işlemleri oldukça kolay bir şekilde yapmanızı sağlar:

- **Toplama:** `+`
- **Çıkarma:** `-`
- **Çarpma:** `*`
- **Bölme:** `/`
- **Tam bölme:** `//` (sonucu tam sayı olarak döndürür)
- **Mod alma (kalan):** `%`
- **Üs alma:** `**`



Sayılarla işlemler yaparken değişkenlere değer atayabilir ve bu değişkenler üzerinde istediğiniz aritmetik işlemleri uygulayabilirsiniz.

```python
a = 5
b = 10.5

print(a * 3) # 15
print(a / 7) # 0.7142857142857143
print(a * b / 10) # 5.25
print(a ** 2) # 25
```

## Tipleri  Değiştirmek

Python, değişkenlerin tiplerini dinamik olarak değiştirebilir. Örneğin, bir değişkenin sayısal bir değer tuttuğunu belirten bir tip bildirimi yapmadan önce, farklı bir tipe dönüştürebilirsiniz.

- `int()`: Bir değeri tam sayıya (integer) dönüştürür.
- `float()`: Bir değeri ondalıklı sayıya (float) dönüştürür.
- `str()`: Bir değeri karakter dizisine (string) dönüştürür.
- `complex()`: Bir değeri karmaşık sayıya (complex number) dönüştürür.


Özellikle, kullanıcıdan alınan veriler genellikle string tipindedir; sayısal işlem yapmak için önce `int()` veya `float()` ile uygun tipe dönüştürebilirsiniz.

```python
int(b)
```

```python
float(a)
```

```python
int(a * b / 10)
```

Fonksiyonlar ve metotlar arasındaki temel fark, fonksiyonların kendi başlarına çağrılabilmeleri,

## Sözlük (Dictionary)
- Sözlükler (dictionary), anahtar (key) ve değer (value) çiftlerinden oluşur.
- Her anahtar benzersizdir ve bir değere karşılık gelir.
- Elemanlara anahtarlarıyla erişilir.
- Sözlükler değiştirilebilir (mutable) yapılardır.
- Sırasızdır, yani elemanların belirli bir sırası yoktur.
- Hızlı veri erişimi ve eşleştirme işlemleri için tercih edilir.
- Anahtarlar, değerler veya anahtar-değer çiftleri özel metodlarla alınabilir.

```python
dictionary = {
    "REG": "Regression",
    "LOG": "Logistic Regression",
    "CART": "Classification and Reg"
}

dictionary["REG"]
```

```python
dictionary = {
    "REG": ["RMSE", 10],
    "LOG": ["MSE", 20],
    "CART": ["SSE", 30]
}
```

```python
"REG" in dictionary # "REG" anahtarı dictionary içinde var mı kontrol eder.
```

```python
dictionary.get("REG") # "REG" anahtarının değerini getirir.
```

```python
dictionary["REG"] = ["RMSE", 100] # "REG" anahtarının değerini değiştirir.
```

```python
dictionary.keys() # dictionary içindeki tüm anahtarları getirir.
```

```python
dictionary.values() # dictionary içindeki tüm değerleri getirir.
```

```python
dictionary.items() # dictionary içindeki tüm anahtar-değer çiftlerini getirir.
```

```python
dictionary.update({"REG": ["RMSE", 100]}) # dictionary içine yeni bir anahtar-değer çifti ekler.
```

```python
dictionary.update({"REG": ["RMSE", 100]}) # dictionary günceller.
```

Tuple, Python'da birden fazla değeri tek bir veri yapısı içerisinde saklamak için kullanılan sıralı (ordered) ve değiştirilemez (immutable) bir veri tipidir.

Tuple'lar köşeli parantez [] yerine normal parantez () ile oluşturulur.

Örnek:
t = (1, 2, 3)

Bir tuple'ın elemanları farklı veri tiplerinden oluşabilir.
Örnek:
t = (1, "hello", 3.5)

Tuple'larda indeksleme yapılabilir.
Örnek:
t[0]  # 1

Tuple'lar değiştirilemez. Eleman ekleme, silme veya değiştirme işlemleri yapılamaz.

Tuple'ın uzunluğu len() fonksiyonu ile bulunabilir.
len(t)  # 3

Tuple içinde bir elemanın olup olmadığı in ile kontrol edilebilir.
1 in t  # True

```python
t = ("john","mark",1,2,3)

t[0]

```

```python
t[0:3]
```

```python
t[0] = "ali" # tuple'ın elemanını değiştirmek istediğimizde hata alırız.
```

## Set (Küme)

- Set, Python'da benzersiz (unique) ve sırasız (unordered) öğeleri saklamak için kullanılan bir veri yapısıdır.
- Set'ler değiştirilebilir ancak elemanları benzersiz olmalıdır yani aynı değerden birden fazla bulunamaz.
- Set'lerde indeksleme ve dilimleme yapılamaz çünkü öğelerin sırası yoktur.
- Eleman ekleme ve çıkarma işlemleri yapılabilir.
- Bir set'in uzunluğu len() fonksiyonu ile bulunabilir ve bir elemanın set içinde olup olmadığı in anahtar kelimesiyle kontrol edilebilir.

```python
set1 = {1,3,5}
set2 = set([1,2,3])

set1.difference(set2) # set1'de set2'de olmayan elemanları döndürür.
```

```python
set2.difference(set1) # set2'de set1'de olmayan elemanları döndürür.
```

```python
set1.symmetric_difference(set2) # set1 ve set2'deki farklı elemanları döndürür.
```

```python
set2.symmetric_difference(set1) # set2 ve set1'deki farklı elemanları döndürür.
```

```python
set1.intersection(set2) # set1 ve set2'deki ortak elemanları döndürür.
```

```python
set1.union(set2) # set1 ve set2'deki tüm elemanları döndürür.
```

```python
set1.isdisjoint(set2) # set1 ve set2'deki elemanların ortak olup olmadığını kontrol eder.
```

```python
set2.issuperset(set1) # set2'nin set1'i kapsadığını kontrol eder.
```
