"""
## List Comprehensions

List comprehensions, Python'da bir listeyi hızlı ve pratik şekilde oluşturmanın kısa yoludur. Sıklıkla bir döngü (for loop) ve opsiyonel olarak koşullu ifadeler (if condition) ile birlikte kullanılırlar. Daha okunaklı ve daha az kod yazarak, yeni bir liste yaratmak için ideal bir yöntemdir.

Temel Söz Dizimi:
```
[expression for item in iterable if condition]
```

### Örnekler

**Örnek 1:** Bir listenin elemanlarının karesini almak:
```
numbers = [1, 2, 3, 4, 5]
squares = [x**2 for x in numbers]
# squares => [1, 4, 9, 16, 25]
```

**Örnek 2:** Bir listeden sadece çift sayıları seçmek:
```
numbers = [1, 2, 3, 4, 5, 6]
evens = [x for x in numbers if x % 2 == 0]
# evens => [2, 4, 6]
```

**Örnek 3:** İki listenin elemanlarını birleştirerek çiftler oluşturmak:
```
colors = ["red", "blue"]
objects = ["car", "pen"]
combinations = [color + " " + obj for color in colors for obj in objects]
# combinations => ['red car', 'red pen', 'blue car', 'blue pen']
```

List comprehensions, kodu daha kompakt ve okunabilir hale getirerek veri işleme işlemlerini kolaylaştırır.
"""

````python
salaries = [1000, 2000, 3000, 4000, 5000]

def new_salary(x):
    return x*20/100 + x

for salary in salaries:
    print(new_salary(salary))

null_list = []

for salary in salaries:
    if salary > 3000:
        null_list.append(new_salary(salary))
    else:
        null_list.append(new_salary(salary * 2))

null_list
````

````python
[new_salary(salary * 2) if salary < 3000 else new_salary(salary) for salary in salaries]
````

````python
[salary * 2 for salary in salaries ]
````

````python
[salary * 2 for salary in salaries if salary < 3000]
````

````python
[salary * 2 if salary < 3000 else salary * 0 for salary in salaries]
````

````python
students = ["John", "Mark", "Venessa", "Mariam"]
students_no = ["John", "Venessa"]

[student.lower() if student in students_no else student.upper() for student in students]



````

"""
Dict comprehension, sözlük (dictionary) nesnelerini hızlı ve pratik bir şekilde oluşturmak için kullanılan bir yapıdır. Liste ve set comprehension yapısına benzer şekilde çalışır, ancak anahtar (key) ve değer (value) çiftleriyle işlem yapılır.

Temel sözdizimi şu şekildedir:

{anahtar_ifadesi: değer_ifadesi for öğe in iterable}

Örnek:

Çift sayıların karesini anahtar ve değer olarak içeren bir sözlük oluşturalım:

squares = {x: x**2 for x in range(5) if x % 2 == 0}

Yukarıdaki örnek sonucunda aşağıdaki sözlük elde edilir:
{0: 0, 2: 4, 4: 16}
"""

````python
dictionary = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}

{k: v ** 2 for k, v in dictionary.items()}
````

````python
{k.upper(): v * 2 for k, v in dictionary.items()}

````

````python
# İsmininde "INS" olan değişkenlerin başına FLAG diğerlerine NO_FLAG eklemek istiyoruz.
df.columns = ["FLAG_" + col if "INS" in col else "NO_FLAG_" + col for col in df.columns]
df.columns
````

````python
# Amaç key'i string, value'su aşağıdaki gibi bir liste olan sözlük oluşturmak.
# 
# Output:
# {'total': ['mean', 'min', 'max', 'var'],
#  'speeding': ['mean', 'min', 'max', 'var'],
#  'alcohol': ['mean', 'min', 'max', 'var'],
#  'not_distracted': ['mean', 'min', 'max', 'var'],
#  'no_previous': ['mean', 'min', 'max', 'var'],
#  'ins_premium': ['mean', 'min', 'max', 'var'],
#  'ins_losses': ['mean', 'min', 'max', 'var']}

import seaborn as sns
df = sns.load_dataset('car_crashes')
df.columns

num_cols = [col for col in df.columns if df[col].dtype != "O"]
soz = {}
agg_list = ['mean', 'min', 'max', 'var']
for col in num_cols:
    soz[col] = agg_list
soz



````

````python
{col:agg_list for col in num_cols}
````
