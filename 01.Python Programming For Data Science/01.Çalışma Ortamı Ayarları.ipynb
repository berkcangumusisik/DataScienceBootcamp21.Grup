### print() Fonksiyonu Nedir?

`print()` fonksiyonu, Python programlarında ekrana (standart çıktı) veri yazdırmak için kullanılan yerleşik bir fonksiyondur. Programcı, ekranda bir bilginin veya hesaplama sonucunun görünmesi gerektiğinde `print()` fonksiyonunu kullanır. İçerisine yazılan ifadeleri, metinleri veya değişkenleri ekrana sıralı şekilde aktarır. Ayrıca, birden fazla öğe yazdırmak istendiğinde bu öğeler aralarına virgül koyularak aynı satırda gösterilebilir. Sonuç olarak, kullanıcıya bilgi aktarmanın en temel yoludur ve Python'da çıktıyı ekrana vermek için en sık başvurulan fonksiyonlardan biridir.


````python
print("Hello World")
````

````python
print("Hello AI Era")
````

#### Sayılar
- Matematiksel işlemler ve hesaplamalarda kullanılan veri türleridir.
- Tam sayılar (integer) ve ondalıklı sayılar (float) olarak ikiye ayrılır.
- Üzerlerinde toplama, çıkarma, çarpma, bölme gibi temel matematiksel işlemler yapılabilir.
- Değişkenler aracılığıyla kullanılabilir ve programlarda farklı amaçlar için saklanabilirler.

#### Karakter Dizileri (Stringler)
- Metinsel ifadeleri saklamak ve işlemek için kullanılır.
- Tek tırnak (‘ ’), çift tırnak (“ ”) veya üçlü tırnak (“”” “””) içinde yazılırlar.
- Birleştirme (concatenation), tekrar etme ve parça alma (slicing) gibi işlemler yapılabilir.
- Metin tabanlı veriyle çalışırken en çok kullanılan veri tiplerinden biridir.

````python
print(9)
````

### type() Fonksiyonu Nedir?

- `type()` fonksiyonu, Python'da bir değişkenin veya ifadenin veri tipini (türünü) öğrenmek için kullanılan yerleşik bir fonksiyondur.
- Kullanıcı, değişkenin veya değerin hangi tipte olduğunu görmek istediğinde `type()` fonksiyonunu çağırır.
- Örneğin, bir sayının int mi yoksa float mı, bir ifadenin string mi olduğunu anlamak için kullanılır.
- Söz dizimi: `type(değer)` şeklindedir ve sonucunda ilgili veri tipini ekrana yazdırır (ör. `<class 'int'>`, `<class 'str'>`).


````python
type(9.2)
````

````python
type("Merhaba")
````

### Atamalar ve Değişkenler

- **Değişkenler**, verileri programda saklamak ve daha sonra kullanmak için verilen isimlerdir.
- Bir değeri bir değişkene atamak için `=` atama operatörü kullanılır.
- Değişken isimleri harf, rakam ve alt çizgi (`_`) içerebilir, ancak rakamla başlayamaz.
- Değişken isminde boşluk ve özel karakterler bulunmaz.
- Atama yapıldığında, değişken bellekte ilgili değeri tutar ve tekrar kullanılabilir.
- Örnek:
  ```python
  x = 10
  isim = "Ali"
  pi = 3.14
  ```
- Değişkenin tipini değiştirmek için yeni bir değer atanabilir.
- Python’da bir değişkenin tipi, atanan değere göre otomatik olarak belirlenir.

````python
a = 9
print(a)
````

````python
b = "Hello Ai Era"
print(b)
````

### Virtual Environment (Sanal Ortam)
- İzole çalışma ortamları oluşturmak için kullanılan araçlardır.
- Farklı çalışmalar için oluşabilecek farklı kütüphane ve versiyon ihtiyaçlarını çalışmalar birbirini etkilemeyecek şekilde oluşturma imkanı sağlar.
- Örneğin; bazı projelerde kullanılan Python, numpy ve pandas sürümleri birbirinden farklı olabilir:
    - Proje 1: python 3.7, numpy 1.16, pandas 1.1
    - Proje 2: python 2.7, numpy 1.14, pandas 0.14
    - Proje 3: python 3.8, numpy 1.20, pandas 0.19
- Eğer sanal ortam kullanılmazsa, bu farklı sürüm ihtiyaçları çakışabilir ve projeler birbirini olumsuz etkileyebilir.
- Sanal ortam sayesinde projeler birbirinden bağımsız olarak çalışır ve her projenin ihtiyaç duyduğu kütüphane ve sürümler güvenle yönetilebilir.
- Kısacası, sanal ortamlar projeler arasında kütüphane ve sürüm çatışmasını engeller ve daha düzenli bir çalışma ortamı sağlar.
#### Örnek:

Diyelim ki, Python’da oyun yapmak için `pygame` kütüphanesini kullanmak istiyoruz.
Ancak bazı kütüphaneler sadece belirli Python sürümleriyle uyumludur.

**pygame der ki:**  
“Ben Python 3 serisi ile çalışamıyorum, sadece 2 ile çalışabiliyorum.”

Bizim sistemimizde ise tüm uygulamalarda Python 3 yüklü ve başka bir Python sürümüne geçmek pratik değil.

Örneğin, sistemimizde kurulu paketler ve sürümleri aşağıdaki gibi olabilir:
```
python        3.7.6
pygame        2.0.0
```

Böyle bir durumda;
- Farklı projeler için farklı Python veya kütüphane sürümlerine ihtiyaç duyulabilir.
- Eğer çalışma ortamlarımızı izole etmezsek, bu tür kütüphane/sürüm uyumsuzlukları proje geliştirmeyi zorlaştırır.

**Çözüm:**  
Sanal ortamlar (virtual environment) kurarak, her projede ihtiyaca yönelik Python ve kütüphane sürümlerini izole kullanabiliriz.

Özetle:  
Sanal ortamlar sayesinde, farklı sürüm gereksinimlerine sahip projeler bir arada sorunsuz şekilde geliştirilebilir ve çalıştırılabilir.

### Sanal Ortam Araçlarının Karşılaştırılması

- venv: Python ile birlikte gelen ve projeler için izole çalışma ortamları oluşturmaya yarayan bir araçtır.
- virtualenv: Standart kütüphanelerden bağımsız olarak çalışan ve farklı Python projeleri için sanal ortam oluşturan bir araçtır.
- pipenv: Sanal ortam yönetimi ve bağımlılık yönetimini bir arada sağlayan modern bir araçtır.
- conda: Hem sanal ortam hem de paket yönetimini sağlayan, özellikle veri bilimi projelerinde yaygın olarak kullanılan bir platformdur.

### Paket Yönetim (Package Management) Araçları

Python projelerinde paketlerin ve sürümlerinin kolayca yönetilmesini sağlayan temel araçlar şunlardır:

- **pip**: Projeye gerekli paketleri kurmak ve bu paketlerin sürümlerini yönetmek için kullanılır. Paket listesi ve sürümleri `requirements.txt` dosyasında tutulur.
- **pipenv**: Paket bağımlılıkları ile birlikte sanal ortamı bir arada yönetir. Bağımlılıklar ve ortam bilgisi `Pipfile` ve `Pipfile.lock` dosyalarında saklanır.
- **conda**: Hem paketleri hem de izole ortamları yönetir; özellikle veri bilimi alanında yaygındır. Ortama ait konfigürasyonlar `environment.yaml` dosyasında belirtilir.

- pip için: `requirements.txt`  
- pipenv için: `Pipfile`  
- conda için: `environment.yaml`  

Bu araçlar, projeye ait tüm bağımlılıkların tekrar kurulmasını ve aynı ortamın kolayca yeniden oluşturulmasını sağlar.

### Sanal Ortam ve Paket Yönetimi Komutları

Verimli ve sürdürülebilir Python projeleri geliştirmek için sanal ortamlar ile paket yönetimi çok önemlidir. Aşağıda yaygın olarak kullanılan komutlar ve açıklamaları yer almaktadır:

#### Conda ile Sanal Ortam ve Paket Yönetimi

- `conda env list`  
  Sistemdeki mevcut conda ortamlarını listeler.

- `conda create -n myenv`  
  `myenv` isimli yeni bir sanal ortam oluşturur.

- `conda activate myenv`  
  `myenv` isimli ortamı aktif hale getirir.

- `conda list`  
  Aktif ortamda yüklü paketleri ve sürümlerini gösterir.

- `conda install numpy`  
  Ortama `numpy` kütüphanesini kurar.

- `conda install numpy scipy pandas`  
  Aynı anda birden çok paketi kurar (ör: `numpy`, `scipy`, `pandas`).

- `conda remove package_name`  
  Belirtilen paketi ortamdan kaldırır.

- `conda install numpy=1.20.1`  
  `numpy` kütüphanesinin 1.20.1 sürümünü kurar.

- `conda upgrade numpy`  
  `numpy` kütüphanesini günceller.

- `conda upgrade --all`  
  Ortamdaki tüm paketleri günceller.

---

#### Pip ile Paket Yönetimi

Python’un varsayılan paket yönetici aracı pip ile temel komutlar şunlardır:

- `pip install pandas`  
  `pandas` kütüphanesini kurar.

- `pip install pandas==1.2.1`  
  `pandas` kütüphanesinin belirli bir (1.2.1) sürümünü kurar.

---

Bu komutlar ile projelerinize özel sanal ortamlar oluşturabilir ve gerekli paketleri/versiyonları rahatça yönetebilirsiniz.
