# Pandas Nedir?

Pandas, Python programlama diliyle veri analizi ve veri işleme işlemlerini kolaylaştırmak için geliştirilmiş açık kaynak kodlu bir kütüphanedir.

Özellikle tablo (DataFrame) ve seri (Series) gibi veri yapıları sunarak, düzenli ve karmaşık veriler üzerinde kolayca işlem yapmayı mümkün kılar.

Pandas ile:

- CSV, Excel gibi dosya tiplerinden veri okuyabilir ve kaydedebilirsiniz.
- Verileri filtreleyebilir, gruplayabilir, özetleyebilir ve dönüştürebilirsiniz.
- Eksik verilerle (NaN) kolayca başa çıkabilirsiniz.
- İstatistiksel analizler ve veri temizleme işlemleri gerçekleştirebilirsiniz.

Pandas, veri bilimi, makine öğrenimi ve istatistiksel analiz gibi alanlarda temel bir araçtır ve veriyle çalışanlar için oldukça gereklidir.

## Pandas Series Nedir?

Pandas `Series`, tek boyutlu (1D) veri tutmak için kullanılan özel bir veri yapısıdır. Bir `Series`; bir veri dizisi ve her veri için bir etiket (index) içerir. Bu sayede, verileri kolayca erişilebilir ve işlenebilir hale getirir.

**Temel Özellikleri:**
- Bir `Series`, Python'un yerleşik listesi veya NumPy dizisi gibi düşünülebilir; fakat her öğenin bir etiketi (index) vardır.
- Veriler farklı veri tiplerinden oluşabilir (tamsayı, float, string vb.).
- Eksik değerlere (NaN) izin verir.

**Series Oluşturma Örneği:**
```python
import pandas as pd

veriler = [10, 20, 30, 40]
etiketler = ['a', 'b', 'c', 'd']

s = pd.Series(veriler, index=etiketler)
print(s)
```
Çıktı:
```
a    10
b    20
c    30
d    40
dtype: int64
```

**Kısaca:**  
Pandas Series, tek boyutlu veri kümelerini etiketlerle birlikte kolayca saklamanızı ve işlemenizi sağlar. Özellikle zaman serisi analizlerinde ve sütun bazlı veri işlemlerinde sıklıkla kullanılır.

````python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
print(df.head()) # ilk 5 veriyi getirir
print("--------------------------------")
print(df.tail()) # son 5 veriyi getirir
print("--------------------------------")
print(df.info()) # veri tiplerini getirir
print("--------------------------------")
print(df.index) # indexleri getirir
print("--------------------------------")
print(df.describe().T) # istatistiksel verileri getirir
print("--------------------------------")
print(df.isnull().values.any()) # eksik verileri getirir
print("--------------------------------")
print(df.isnull().sum()) # eksik verilerin sayısını getirir
print("--------------------------------")

````
