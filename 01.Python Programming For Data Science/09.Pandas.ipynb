## Pandas Series Nedir?

Pandas `Series`, tek boyutlu (1D) veri tutmak için kullanılan özel bir veri yapısıdır. Bir `Series`; bir veri dizisi ve her veri için bir etiket (index) içerir. Bu sayede, verileri kolayca erişilebilir ve işlenebilir hale getirir.

**Temel Özellikleri:**
- Bir `Series`, Python'un yerleşik listesi veya NumPy dizisi gibi düşünülebilir; fakat her öğenin bir etiketi (index) vardır.
- Veriler farklı veri tiplerinden oluşabilir (tamsayı, float, string vb.).
- Eksik değerlere (NaN) izin verir.

**Series Oluşturma Örneği:**
```python
import pandas as pd

veriler = [10, 20, 30, 40]
etiketler = ['a', 'b', 'c', 'd']

s = pd.Series(veriler, index=etiketler)
print(s)
```
Çıktı:
```
a    10
b    20
c    30
d    40
dtype: int64
```

## Pandas'ta Seçim İşlemleri (Selection in Pandas)

Pandas'ta veri seçme işlemleri farklı yöntemlerle yapılabilir. En yaygın seçim işlemleri şunlardır:

**1. Tek bir sütunu seçmek:**  
Bir DataFrame'den bir sütun seçmek için köşeli parantezler kullanılır:  
`df['column_name']`  
veya  
`df.column_name`

**2. Birden fazla sütunu seçmek:**  
Birden fazla sütun seçmek için sütun isimlerini bir liste içinde belirtirsiniz:  
`df[['column1', 'column2']]`

**3. Satıra göre seçim (loc ve iloc):**  
- `loc[]` etikete (index adı) göre seçim yapar:  
  `df.loc[2]` → indexi 2 olan satırı getirir  
  `df.loc[2:5]` → 2’den 5’e kadar (5 dahil) satırları getirir  
  `df.loc[:, 'column_name']` → tüm satırlarda sadece belirtilen sütunu getirir

- `iloc[]` pozisyona göre seçim yapar (0’dan başlar):  
  `df.iloc[0]` → ilk satırı getirir  
  `df.iloc[0:3]` → 0, 1 ve 2. satırları getirir  
  `df.iloc[:, 1]` → tüm satırlarda 2. sütunu getirir

**4. Koşullu seçim:**  
Belirli bir koşulu sağlayan satırları seçmek için boolean koşullar kullanılabilir:  
`df[df['age'] > 18]` → Yaşı 18’den büyük olanları seçer

**5. Belirli hücreye erişmek:**  
Belirli bir satır ve sütundaki değere ulaşmak için:  
`df.loc[3, 'age']` veya `df.iloc[3, 2]` kullanılabilir.

Bu yöntemler sayesinde Pandas'ta veri seçimini esnek ve kolay bir şekilde gerçekleştirebilirsiniz.

## `.loc` ve `.iloc` Kullanımı

Pandas DataFrame'lerinde satır ve sütun seçimleri için iki temel özellik kullanılır: `.loc` ve `.iloc`.

### `.loc` (Etiket/İsim Bazlı Seçim)

- Satır ve sütunlara etiket (isim) ile erişim sağlar.
- Söz dizimi: `df.loc[satır_seçimi, sütun_seçimi]`
- Hem satırda hem sütunda tek bir isim, liste veya aralık kullanılabilir.
- Örnekler:
    - `df.loc[5]`: İndeksi 5 olan satırı getirir.
    - `df.loc[2:6]`: 2’den 6’ya kadar olan satırları getirir (**son değer dahil**).
    - `df.loc[:, "age"]`: Tüm satırlarda "age" sütununu getirir.
    - `df.loc[0:3, ["age", "fare"]]`: 0’dan 3’e kadar olan satırların "age" ve "fare" sütunlarını getirir.

### `.iloc` (Pozisyon/İndis Bazlı Seçim)

- Satır ve sütunlara indis (integer konumu) ile erişim sağlar.
- Söz dizimi: `df.iloc[satır_seçimi, sütun_seçimi]`
- Seçim 0 tabanlıdır; aralıklarda üst sınır **hariçtir**.
- Örnekler:
    - `df.iloc[0]`: İlk satırı getirir.
    - `df.iloc[2:5]`: 2, 3, 4. satırları getirir.
    - `df.iloc[:, 1]`: Tüm satırlarda, 2. sütunu (indis 1) getirir.
    - `df.iloc[0:3, [2, 4]]`: 0, 1, 2. satırdan; 2. ve 4. sütunları getirir.

### Temel Farklar

| Özellik               | `.loc`             | `.iloc`                  |
|-----------------------|--------------------|--------------------------|
| Erişim tipi           | Etiket/İsim        | Sayısal Pozisyon         |
| Aralık sistemi        | Son dahil          | Son hariç                |
| Hata davranışı        | Etiket yoksa hata  | Sınır dışıysa hata       |

`.loc` ve `.iloc`'u karıştırmadan, ihtiyacınıza göre veriye doğrudan erişebilirsiniz.

Koşullu seçimler, pandas DataFrame’lerinde belirli şartları sağlayan verileri filtrelemek için kullanılır. Bir sütunda veya birden fazla sütunda koşul belirlenerek sadece bu koşulu sağlayan satırlar seçilebilir. Tüm şartlarda, her bir koşul parantez içine alınmalıdır. Koşullu seçimler veri analizi sürecinde ilgilenilen spesifik verileri elde etmek için oldukça etkilidir.

### Toplulaştırma (Aggregation) ve Gruplama (Grouping)

Veri analizinde, büyük veri setlerinde alt gruplar oluşturmak ve bu gruplar üzerinde toplu özet istatistikler (ortalama, toplam, adet vb.) almak oldukça önemlidir. Pandas kütüphanesinde bu işlemler oldukça pratik bir şekilde yapılabilir.

#### **Gruplama (groupby) Nedir?**
Bir DataFrame'i belirli bir sütuna göre böler ve her benzersiz kategori için ayrı birer alt tablo (grup) oluşturur. Böylece her grup üzerinde ayrı analizler ve hesaplamalar yapmak mümkün olur.

#### **Toplulaştırma (aggregation) Nedir?**
Gruplanmış alt veri kümeleri üzerinde çeşitli özet istatistikleri (ör. ortalama, toplam, adet, maksimum, minimum) hesaplamak için kullanılır.

#### **En Sık Kullanılan groupby ve aggregate Fonksiyonları**
- Eleman sayısı (`count()`)  
  Her grup içindeki elemanların sayısını verir.
- İlk değer (`first()`)  
  Her grup için ilk veriyi döndürür.
- Son değer (`last()`)  
  Her grup için son veriyi döndürür.
- Ortalama (`mean()`)  
  Sayısal değerlerin ortalamasını alır.
- Medyan (`median()`)  
  Ortanca değeri döndürür.
- En küçük değer (`min()`)  
  Her grup için minimumu döndürür.
- En büyük değer (`max()`)  
  Her grup için maksimumu döndürür.
- Standart sapma (`std()`)  
  Her grup için standart sapmayı döndürür.
- Birden fazla istatistik `agg()` ile aynı anda uygulanabilir.

Not: Yukarıdaki metotlar, pandas gruplama işlemlerinde en çok kullanılan toplulaştırma fonksiyonlarındandır.

#### **Kullanım Örnekleri**
```python
# Sınıfa göre ortalama yaş:
df.groupby("class")["age"].mean()

# Cinsiyete göre toplam bilet ücreti:
df.groupby("sex")["fare"].sum()

# Sınıf ve cinsiyete göre ortalama hayatta kalma oranı:
df.groupby(["class", "sex"])["survived"].mean()

# Sınıfa göre yaşın ortalaması ve bilet ücretinin toplamı, maksimumu:
df.groupby("class").agg({"age": "mean", "fare": ["sum", "max"]})

# Sınıfa göre yaşın ortalaması, medyanı ve standart sapması:
df.groupby("class")["age"].agg(["mean", "median", "std"])
```

#### **Neden Önemlidir?**
Gruplama ve toplulaştırma işlemleri, veri setlerinde örüntüleri, ilişkileri ve özet bilgileri hızlıca görmeyi sağlar. Analitik kararlar almak ve raporlar üretmek için güçlü ve vazgeçilmez araçlardır.

````python
df["age"].mean() # yaşın ortalamasını getirir
````

````python
df.groupby("sex")["age"].mean() # cinsiyete göre yaşın ortalamasını getirir
````

````python
df.groupby("sex").agg({"age": "mean"}) # cinsiyete göre yaşın ortalamasını getirir
````

````python
df.groupby("sex").agg({"age": ["mean","sum"]})
````


## Pivot Table (Özet Tablo) Nedir?

Pivot table, veri kümesindeki bilgileri farklı boyutlarda özetleyerek tablo halinde gösterebilen güçlü bir araçtır. Pandas kütüphanesindeki `pivot_table` fonksiyonu ile bir DataFrame'den kolayca özet tablolar oluşturulabilir. Böylece, satır ve sütun başlıklarını istediğimiz gruplara göre ayarlayabilir ve çeşitli toplulaştırma (aggregation) fonksiyonlarını aynı anda uygulayabiliriz.

Örneğin, Titanic veri setinde yolcuların cinsiyetine, bindiği limana ve sınıfına göre ortalama yaş, hayatta kalma oranı ve yolcu sayısını özetlemek için şöyle bir pivot tablo oluşturulabilir:

```python
df.pivot_table(
    values=["age", "survived", "sex"],
    index=["sex", "embark_town", "class"],
    aggfunc={"age": "mean", "survived": "mean", "sex": "count"}
)
```

Burada:
- `values`: Hesaplanacak öznitelikleri belirtir (ör. yaş, hayatta kalma, cinsiyet sayısı)
- `index`: Satır başlıklarını belirler (ör. cinsiyet, biniş limanı, sınıf)
- `aggfunc`: Her sütunda hangi toplulaştırma fonksiyonunun uygulanacağını belirler

Sonuç olarak oluşacak tablo, satırda seçilen gruplara göre her bir grubun yaş ortalaması, hayatta kalma oranı ve toplam yolcu sayısını gösterir. Bu şekilde çok boyutlu veri analizi ve raporlama kolayca yapılabilir.



````python
df.pivot_table("survived","sex","embarked")
````

````python
df.pivot_table("survived","sex","embarked", aggfunc="std")
````

````python
df.pivot_table("survived","sex",["embarked","class"])
````

````python
df["new_age"] = pd.cut(df["age"], [0,10,18,25,40,90])
df.head()
# pd.cut fonksiyonu, bir sütunu belirli aralıklara bölerek yeni bir sütun oluşturur. 
# Örneğin, df["age"] sütununu 0-10, 10-18, 18-25, 25-40, 40-90 aralıklarına böler ve yeni bir sütun oluşturur.
````

### `apply` ve `lambda` Fonksiyonları

Pandas'ta `.apply()` metodu, bir DataFrame veya Series üzerindeki satırlara ya da sütunlara özel bir fonksiyonu uygulamak için kullanılır. Özellikle hızlı işlemler, dönüşümler veya karmaşık fonksiyonlar için oldukça faydalıdır.

#### Lambda Fonksiyonu
`lambda`, Python'da kısa ve isimsiz (anonim) fonksiyonlar tanımlamak için kullanılır. Genelde küçük, tek satırlık işlemler için tercih edilir.

#### Örnek Kullanım

```python
import pandas as pd

df = pd.DataFrame({
    "sayi": [1, 2, 3, 4, 5]
})

# 'sayi' sütununa, her değerin karesini alan yeni bir sütun ekleyelim:
df["karesi"] = df["sayi"].apply(lambda x: x**2)
print(df)
```

Yukarıdaki örnekte:
- `df["sayi"].apply(lambda x: x**2)` ifadesiyle, "sayi" sütunundaki her bir değere `lambda x: x**2` fonksiyonu uygulanır.
- Sonuç olarak "karesi" adında, orijinal değerlerin karesi bulunan yeni bir sütun elde edilir.

`.apply()` metodu, satır ya da sütun bazında işlem yapmak ve özelleştirilmiş fonksiyonlar tanımlamak için oldukça kullanışlıdır.

````python
for col in df.columns:
    if "age" in col:
        print(col)
````

````python
for col in df.columns:
    if "age" in col:
        if pd.api.types.is_numeric_dtype(df[col]):
            print((df[col]/10).head())
        else:
            print(f"{col} sütunu sayısal olmadığı için bölme işlemi uygulanamadı.")
   
````

## Birleştirme (Merge) İşlemleri

Pandas'ta birleştirme (merge), iki veya daha fazla veri çerçevesini (DataFrame) anahtar sütunlar üzerinden bir araya getirmek için kullanılır. Farklı veri kaynaklarındaki bilgileri tek bir yapıda toplamak için çok önemli bir işlemdir.

### Temel Birleştirme Yöntemi: `pd.merge()`

İki DataFrame'i birleştirmek için `pd.merge()` fonksiyonu kullanılır. Temel kullanım şekli:

```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'ad': ['Ali', 'Veli', 'Ayşe', 'Fatma']
})

df2 = pd.DataFrame({
    'id': [1, 2, 5, 6],
    'puan': [95, 80, 75, 88]
})

# id sütunu üzerinden birleştirme (inner join)
birlesik_df = pd.merge(df1, df2, on='id')
```

### Birleştirme Türleri

`how` parametresiyle farklı birleştirme türleri seçilebilir:

- **inner**: Ortak olan anahtarları getirir (kesişim).
- **left**: Birinci DataFrame'in tüm satırlarını getirir, eşleşmeyenlere NaN yazar.
- **right**: İkinci DataFrame'in tüm satırlarını getirir, eşleşmeyenlere NaN yazar.
- **outer**: Tüm anahtarları getirir (birleşim).

```python
# 'left' join örneği
pd.merge(df1, df2, on='id', how='left')

# 'outer' join örneği
pd.merge(df1, df2, on='id', how='outer')
```

### Farklı Sütun İsimleriyle Birleştirme

Eğer anahtar sütunlarının isimleri farklıysa `left_on` ve `right_on` parametreleri kullanılır.

```python
df3 = pd.DataFrame({'ogrenci_id': [1, 2, 3], 'yas': [22, 21, 20]})

# df1'in 'id' sütunu ile df3'ün 'ogrenci_id' sütunu üzerinden birleştirme
pd.merge(df1, df3, left_on='id', right_on='ogrenci_id')
```

### Çoklu Sütuna Göre Birleştirme

Birden fazla sütuna göre de birleştirme yapılabilir:

```python
pd.merge(df1, df2, on=['id', 'ad'])
```

### Özet

Birleştirme işlemleriyle birden fazla veri kaynağındaki ilişkili bilgileri bir araya getirebilir, zengin ve analiz edilebilir tablolar oluşturabilirsiniz. Doğru anahtarlar ve uygun birleştirme türü seçimi, sağlıklı veri analizleri için kritik önemdedir.
