## RFM Nedir?

**RFM**, müşteri değerini ve müşteri segmentasyonunu anlamak için yaygın olarak kullanılan bir analiz yöntemidir. RFM kısaltması, üç İngilizce kelimenin baş harflerinden oluşur:

- **R (Recency / Yenilik):** Müşterinin en son ne zaman alışveriş yaptığını gösterir. Son alışveriş tarihi ne kadar yakınsa, müşteri o kadar “sıcak” bir müşteridir.
- **F (Frequency / Sıklık):** Müşterinin belirli bir zaman aralığında kaç kere alışveriş yaptığıdır. Sık alışveriş yapan müşteriler, markaya daha bağlı kabul edilir.
- **M (Monetary / Parasal Değer):** Müşterinin toplam harcaması, yani şirkete sağladığı parasal değerdir. Yüksek harcamalar yapan müşterilerin işletmeye katkısı yüksektir.

### RFM Analizinin Amacı

RFM analizi ile müşteriler; son alışveriş zamanı, alışveriş yapma sıklığı ve harcama miktarına göre farklı segmentlere ayrılır. Böylece;

- En değerli müşteriler tespit edilir,
- Yeniden pazarlama ve kampanya stratejileri geliştirilir,
- Müşteri sadakati ve karlılığı artırmaya yönelik aksiyonlar alınır.

### RFM Özetle

| RFM Boyutu | Açıklama                                 |
|------------|------------------------------------------|
| Recency    | Son alışverişin üzerinden geçen süre     |
| Frequency  | Toplam alışveriş yapma sayısı            |
| Monetary   | Toplam harcama miktarı                   |

RFM analizi, pazarlama faaliyetlerinde segmente özel, daha etkili stratejiler uygulanmasına olanak tanır.

## RFM Skorlarına Göre Müşteri Segmentasyonu Nasıl Yapılır?

RFM analizi sonucunda her müşteri için ayrı ayrı R, F ve M skorları hesaplanır. Bu skorlar genellikle 1 (en düşük) ile 5 (en yüksek) arasında puanlanır.

Müşteriler, bu üç skorun kombinasyonuna göre belirli segmentlere ayrılır. Örneğin:

- **En iyi müşteriler:** Yüksek Recency (yakında alışveriş yapmış), yüksek Frequency (sık alışveriş yapan) ve yüksek Monetary (çok harcayan) skorlarına sahip olanlardır.
- **Sadık müşteriler:** Yüksek Frequency ve Monetary, fakat Recency skoru biraz daha düşük olabilir.
- **Geri kazanılması gerekenler:** Eskiden sık ve çok harcama yapan ama son dönemde alışveriş yapmamış müşteriler.
- **Yeni müşteriler:** Yüksek Recency, ancak Frequency ve Monetary skorları düşük olanlar.
- **Potansiyel müşteri adayları:** Ortalama skorları olanlar; kampanya ile daha üst segmente taşınabilirler.

Segmentasyonda yaygın bir yöntem, RFM skorlarının birleştirilerek (örneğin “555” veya “355” gibi) segmentlere gruplanmasıdır.

**Örnek Segmentasyon Kuralları:**
| RFM Skoru      | Segment Adı                 |
|--------------- |----------------------------|
| 555, 554       | En Değerli Müşteriler      |
| 455, 454, 355  | Sadık Müşteriler           |
| 551, 511, 451  | Geri Kazanılacaklar        |
| 155, 154, 114  | Yeni Müşteriler            |
| Diğer          | Geliştirilecek Müşteriler  |

Böylece her bir segmente farklı pazarlama stratejileri ve teklifler tasarlanabilir.

### 4. RFM Metriklerinin Hesaplanması (Calculating RFM Metrics)


```python
df["InvoiceDate"].max()
```

```python
today_date = dt.datetime(2010, 12, 11)
type(today_date)
```

```python
# RFM analizi için müşteri bazında Recency (yenilik), Frequency (alışveriş sıklığı), Monetary (toplam harcama) metrikleri hesaplanıyor:
# - Recency: Müşterinin en son alışverişinin üzerinden kaç gün geçti? (today_date - son alışveriş tarihi)
# - Frequency: Müşteri toplam kaç farklı fatura (alışveriş) oluşturmuş?
# - Monetary: Müşterinin toplam harcama tutarı nedir?
rfm = df.groupby('Customer ID').agg({
    'InvoiceDate': lambda InvoiceDate: (today_date - InvoiceDate.max()).days, # Recency
    'Invoice': lambda Invoice: Invoice.nunique(),                             # Frequency
    'TotalPrice': lambda TotalPrice: TotalPrice.sum()                         # Monetary
})
rfm.head()  # Hesaplanan RFM metriklerinden örnek ilk 5 müşteri kaydı gösteriliyor
```

```python
rfm.columns = ['recency', 'frequency', 'monetary']

rfm.describe().T
```

```python
rfm = rfm[rfm["monetary"] > 0]
rfm.shape
```

##  5. RFM Skorlarının Hesaplanması (Calculating RFM Scores)

```python
# RFM skorlarının hesaplanma sürecini adım adım anlatayım:

# 1. Recency Score: Müşterilerin recency (yenilik) değerleri küçükten büyüğe sıralanır ve 5 gruba bölünür.
#    En güncel (yakın zamanda alışveriş yapmış) müşteriler 5 ile, en eski müşteriler ise 1 ile etiketlenir.
rfm["recency_score"] = pd.qcut(rfm["recency"], 5, labels=[5, 4, 3, 2, 1])

# 2. Frequency Score: Frequency yani alışveriş sıklığı, 'rank' yöntemiyle sıralanıp tekrar 5 gruba ayrılır.
#    En az alışveriş yapanlar 1 puan, en çok alışveriş yapanlar 5 puan alır.
rfm["frequency_score"] = pd.qcut(rfm["frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])

# 3. Monetary Score: Monetary yani toplam harcama değeri yine 'rank' ile sıralanarak 5 eşit parçaya ayrılır.
#    En az harcayanlar 1, en çok harcayanlar 5 puan alır.
rfm["monetary_score"] = pd.qcut(rfm["monetary"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])

# 4. RFM_SCORE'un oluşturulması: Her bir müşteri için recency, frequency ve monetary score değerleri string olarak birleştirilir.
#    Yani örnek bir müşteri için skor örneği: '542' (5: en güncel, 4: sık alışveriş, 2: ortalama harcama gibi)
rfm["RFM_SCORE"] = (
    rfm["recency_score"].astype(str) +
    rfm["frequency_score"].astype(str) +
    rfm["monetary_score"].astype(str)
)
               
```

### 7. Tüm Sürecin Fonksiyonlaştırılması

```python
def create_rfm(dataframe, csv=False):

    # VERIYI HAZIRLAMA
    dataframe["TotalPrice"] = dataframe["Quantity"] * dataframe["Price"]
    dataframe.dropna(inplace=True)
    dataframe = dataframe[~dataframe["Invoice"].str.contains("C", na=False)]

    # RFM METRIKLERININ HESAPLANMASI
    today_date = dt.datetime(2011, 12, 11)
    rfm = dataframe.groupby('Customer ID').agg({'InvoiceDate': lambda date: (today_date - date.max()).days,
                                                'Invoice': lambda num: num.nunique(),
                                                "TotalPrice": lambda price: price.sum()})
    rfm.columns = ['recency', 'frequency', "monetary"]
    rfm = rfm[(rfm['monetary'] > 0)]

    # RFM SKORLARININ HESAPLANMASI
    rfm["recency_score"] = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])
    rfm["frequency_score"] = pd.qcut(rfm["frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
    rfm["monetary_score"] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])

    # cltv_df skorları kategorik değere dönüştürülüp df'e eklendi
    rfm["RFM_SCORE"] = (rfm['recency_score'].astype(str) +
                        rfm['frequency_score'].astype(str))


    # SEGMENTLERIN ISIMLENDIRILMESI
    seg_map = {
        r'[1-2][1-2]': 'hibernating',
        r'[1-2][3-4]': 'at_risk',
        r'[1-2]5': 'cant_loose',
        r'3[1-2]': 'about_to_sleep',
        r'33': 'need_attention',
        r'[3-4][4-5]': 'loyal_customers',
        r'41': 'promising',
        r'51': 'new_customers',
        r'[4-5][2-3]': 'potential_loyalists',
        r'5[4-5]': 'champions'
    }

    rfm['segment'] = rfm['RFM_SCORE'].replace(seg_map, regex=True)
    rfm = rfm[["recency", "frequency", "monetary", "segment"]]
    rfm.index = rfm.index.astype(int)

    if csv:
        rfm.to_csv("rfm.csv")

    return rfm

df = df_.copy()

rfm_new = create_rfm(df, csv=True)
```
