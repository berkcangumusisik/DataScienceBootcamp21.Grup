## Müşteri Yaşam Boyu Değeri (Customer Lifetime Value - CLV) Nedir?

Müşteri Yaşam Boyu Değeri (CLV), bir müşterinin bir işletme ile olan ilişkisi süresince oluşturacağı toplam net kazancı ifade eden bir metriktir. Bir müşteriden elde edilen gelir ile bu müşteriye sunulan ürün ve hizmetler için yapılan maliyetlerin farkı olarak hesaplanır.

CLV, işletmelerin mevcut ve potansiyel müşterilerini daha iyi tanımlamasına, pazarlama ve satış stratejilerini optimize etmesine yardımcı olur. Özellikle müşteri kazanma ve elde tutma maliyetlerini karşılaştırmada önemli bir göstergedir.

Yüksek CLV'ye sahip müşteriler, şirketin uzun vadeli karlılığı için kritik öneme sahiptir. Böyle müşterilere yönelik özel stratejiler geliştirmek, müşteri sadakatini ve kârlılığı artırır.

Müşteri Yaşam Boyu Değeri (CLV) hesaplaması için farklı yöntemler kullanılabilir. Aşağıda hem temel hem daha gelişmiş CLV formülleri adım adım ve detaylı olarak açıklanmıştır.

## Temel CLV Hesaplama Adımları

CLV, genellikle üç ana bileşene dayanır:
1. Ortalama Satış Değeri
2. Satın Alma Sıklığı
3. Müşteri Yaşam Süresi

**Temel CLV formülü:**
```
CLV = Ortalama Satış Değeri x Satın Alma Sıklığı x Müşteri Yaşam Süresi
```

- **Ortalama Satış Değeri:** Bir müşterinin bir alışverişte harcadığı ortalama tutardır.
- **Satın Alma Sıklığı:** Müşterinin belirli bir zaman diliminde ortalama kaç kez alışveriş yaptığıdır.
- **Müşteri Yaşam Süresi:** Müşterinin işletmeyle ilişkisinin ortalama süresidir (genellikle yıl olarak alınır).

## Gelişmiş CLV Hesaplama Adımları

Daha gelişmiş CLV hesaplamalarında müşteri değeri, terk oranı (churn rate) ve kâr oranı (profit margin) dikkate alınmaktadır. Adım adım hesaplama aşağıdadır:

1. **Average Order Value (Ortalama Sipariş Değeri):**
   ```
   Average Order Value = Total Price / Total Transaction
   ```
   Toplam harcama tutarının toplam işlem sayısına bölünmesiyle bulunur.

2. **Purchase Frequency (Satın Alma Sıklığı):**
   ```
   Purchase Frequency = Total Transaction / Total Number of Customers
   ```
   Toplam işlem sayısı, toplam müşteri sayısına bölünür.

3. **Customer Value (Müşteri Değeri):**
   ```
   Customer Value = Average Order Value * Purchase Frequency
   ```

4. **Repeat Rate (Tekrar Alım Oranı):**
   ```
   Repeat Rate = Birden fazla alışveriş yapan müşteri sayısı / Tüm müşteriler
   ```
   Yani tekrar alışveriş yapan müşterilerin oranı hesaplanır.

5. **Churn Rate (Terk Oranı):**
   ```
   Churn Rate = 1 - Repeat Rate
   ```
   Terk oranı, müşterilerin tekrar alışveriş yapmama oranıdır.

6. **Profit Margin (Kâr Marjı):**
   ```
   Profit Margin = Total Price * 0.10
   ```
   (Burada kâr marjı %10 olarak alınmıştır.)

7. **Gelişmiş CLTV formülü:**
   ```
   CLTV = (Customer Value / Churn Rate) x Profit Margin
   ```
   Yani müşteri değeri terk oranına bölünür ve ardından kâr marjı ile çarpılır.
   
Alternatif olarak, gelecekteki kazançları bugüne indirgemek için iskonto oranı kullanılarak daha ileri bir CLV hesaplaması yapılabilir:

```
CLV = (Müşteriden elde edilen yıllık kâr x Müşteri yaşam süresi) / (1 + iskonto oranı) ^ yıl
```

Tüm bu hesaplamalar, firmaların müşteri segmentlerine özel stratejiler geliştirmesine, pazarlama yatırımlarını optimize etmesine ve müşteri kârlılığını maksimize etmesine imkân tanır.


### 1. Veri Hazırlama

Veri seti hikayesi: Online Retail II veri seti, İngiltere merkezli bir online satış mağazasının 01.12.2009 - 09.12.2011 tarihleri arasındaki satışlarını içerir.
Veri setinin kolonları şunlardır:
- InvoiceNo: Her bir fatura (işlem) için eşsiz numara. "C" ile başlıyorsa iptal edilen işlem anlamına gelir.
- StockCode: Her ürünün kendine ait eşsiz kodu.
- Description: Ürün adı.
- Quantity: Satılan ürün adedi (faturada kaç adet satıldığını belirtir).
- InvoiceDate: Faturanın tarihi ve zamanı.
- UnitPrice: Ürün fiyatı, Sterlin cinsinden.
- CustomerID: Eşsiz müşteri numarası.
- Country: Müşterinin bulunduğu ülke.

````python
import pandas as pd

from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_columns', None) # Bu kod, DataFrame'deki tüm sütunları görüntülemek için kullanılır.
# pd.set_option('display.max_rows', None) # Bu kod, DataFrame'deki tüm satırları görüntülemek için kullanılır.
pd.set_option('display.float_format', lambda x: '%.5f' % x) # Bu kod, DataFrame'deki sayıların virgülden sonra kaç basamak görüntüleneceğini belirler.

df_ = pd.read_excel("datasets/online_retail_II.xlsx", sheet_name="Year 2009-2010")

df = df_.copy()
````

````python
df.head()
````

````python
df.isnull().sum() # Bu kod, DataFrame'deki boş değerlerin sayısını hesaplar.
````

````python
df = df[~df["Invoice"].str.contains("C", na=False)] # Bu kod, "C" ile başlamayan satırları filtreler.
````

````python
df.describe().T
````

````python
# Negatif adetli satışları (örneğin iadeleri) çıkarıyoruz; sadece miktarı sıfırdan büyük olan satırlar kalıyor.
df = df[(df['Quantity'] > 0)]

# Eksik (NaN) olan tüm satırları veri setinden siliyoruz.
df.dropna(inplace=True)
````

````python
df["TotalPrice"] = df["Quantity"] * df["Price"]
df
````

````python
cltv_c = df.groupby('Customer ID').agg({'Invoice': lambda x: x.nunique(),
                                        'Quantity': lambda x: x.sum(),
                                        'TotalPrice': lambda x: x.sum()})

cltv_c
````

````python
cltv_c.columns = ['total_transaction', 'total_unit', 'total_price']
````

### 2. Average Order Value (average_order_value = total_price / total_transaction)

````python
cltv_c.head()
````

````python
cltv_c["average_order_value"] = cltv_c["total_price"] / cltv_c["total_transaction"]
````

````python
cltv_c.head()
````

### 3. Purchase Frequency (total_transaction / total_number_of_customers)


````python
cltv_c["total_transaction"]
````

````python
cltv_c.shape[0]
````

````python
cltv_c["purchase_frequency"] = cltv_c["total_transaction"] / cltv_c.shape[0]
cltv_c.head()
````

### 4. Repeat Rate & Churn Rate (birden fazla alışveriş yapan müşteri sayısı / tüm müşteriler)


````python
# Birden fazla alışveriş yapan müşterilerin oranını hesaplar.
repeat_rate = cltv_c[cltv_c["total_transaction"] > 1].shape[0] / cltv_c.shape[0]
repeat_rate
````

````python
# Churn rate'in nasıl hesaplandığını adım adım görelim.

# 1. Önce repeat_rate'i zaten yukarıda hesaplamıştık.
# 2. Churn rate, müşteri kayıp oranı demektir ve 1 - repeat_rate formülüyle bulunur.
# 3. Hesapladıktan sonra sonucu görebilmek için churn_rate'in değerini yazdıralım.

churn_rate = 1 - repeat_rate  # 1'den repeat_rate'i çıkarıyoruz.
churn_rate  # Hesaplanan churn rate'i yazdırıyoruz.
````

### 5. Profit Margin (profit_margin =  total_price * 0.10)

````python
cltv_c['profit_margin'] = cltv_c['total_price'] * 0.10
cltv_c.head()
````

### 6. Customer Value (customer_value = average_order_value * purchase_frequency)

````python
# Customer value (müşteri değeri), her bir müşterinin ortalama sipariş değeri ile satın alma sıklığının çarpımı ile hesaplanır.
# Yani bir müşterinin şirkete sağlayabileceği değeri gösterir.
cltv_c['customer_value'] = cltv_c['average_order_value'] * cltv_c["purchase_frequency"]
cltv_c.head()
````

### 7. Customer Lifetime Value (CLTV = (customer_value / churn_rate) x profit_margin)

````python
cltv_c["cltv"] = (cltv_c["customer_value"] / churn_rate) * cltv_c["profit_margin"]
# Bu satır, 'cltv' (Customer Lifetime Value) değerine göre müşteri tablosunu azalan şekilde sıralar 
# ve en yüksek CLTV'ye sahip ilk 5 müşteriyi gösterir.
cltv_c.sort_values(by="cltv", ascending=False).head()
````

### 8. Segmentlerin Oluşturulması

````python
cltv_c.sort_values(by="cltv", ascending=False).tail()
````

````python
# CLTV (Customer Lifetime Value) skoruna göre müşterileri 4 segmente ayırıyoruz:
# En yüksekten en düşüğe A, B, C, D etiketleri veriliyor; A segmenti en değerli müşterileri temsil eder.
cltv_c["segment"] = pd.qcut(cltv_c["cltv"], 4, labels=["D", "C", "B", "A"])

# CLTV değerine göre sıraladığımızda en yüksek CLTV'ye sahip ilk 5 müşteriyi görebiliriz.
cltv_c.sort_values(by="cltv", ascending=False).head()
````

````python
cltv_c.groupby("segment").agg({"count", "mean", "sum"})
````

###  9. BONUS: Tüm İşlemlerin Fonksiyonlaştırılması

````python
def create_cltv_c(dataframe, profit=0.10):
    """
    Müşteri Yaşam Boyu Değeri (CLTV) hesaplamasını adım adım gerçekleştiren ve segmentlere ayıran fonksiyon.

    Parametreler
    ------------
    dataframe : pd.DataFrame
        Online perakende verisinin yer aldığı ve 'Invoice', 'Quantity', 'Price', 'Customer ID' gibi sütunları içeren veri çerçevesi.
    profit : float, opsiyonel (varsayılan=0.10)
        Kâr marjını belirtir. Tüm satışların % kaçı kâr olarak kabul edilecekse o oran girilir.

    Dönüş
    -----
    cltv_c : pd.DataFrame
        Her müşteri için toplam işlem, toplam adet, toplam harcama, ortalama işlem tutarı, alışveriş sıklığı,
        tekrar alışveriş oranı, churn rate, kâr marjı, müşteri değeri, CLTV ve segmentleri içeren tablo.
    """

    # 1. Faturadan iade (C ile başlayan Invoice) kayıtları çıkarılır.
    dataframe = dataframe[~dataframe["Invoice"].str.contains("C", na=False)]

    # 2. Sıfırdan küçük/negatif miktarlar ve eksik değerler (NA) temizlenir.
    dataframe = dataframe[(dataframe['Quantity'] > 0)]
    dataframe.dropna(inplace=True)

    # 3. Her işlem için toplam ödemenin bulunduğu yeni bir sütun eklenir.
    dataframe["TotalPrice"] = dataframe["Quantity"] * dataframe["Price"]

    # 4. Müşteri bazında özet tablo hazırlanır:
    #    - Toplam fatura adedi (nunique ile farklı fatura sayısı)
    #    - Toplam ürün adedi (satın alman ürünlerin toplamı)
    #    - Toplam harcama tutarı
    cltv_c = dataframe.groupby('Customer ID').agg({
        'Invoice': lambda x: x.nunique(),
        'Quantity': lambda x: x.sum(),
        'TotalPrice': lambda x: x.sum()
    })
    cltv_c.columns = ['total_transaction', 'total_unit', 'total_price']

    # 5. Ortalama Sipariş Değeri (Average Order Value)
    cltv_c['avg_order_value'] = cltv_c['total_price'] / cltv_c['total_transaction']

    # 6. Satın Alma Sıklığı (Purchase Frequency)
    cltv_c["purchase_frequency"] = cltv_c['total_transaction'] / cltv_c.shape[0]

    # 7. Tekrar Alışveriş Oranı ve Churn Rate
    repeat_rate = cltv_c[cltv_c.total_transaction > 1].shape[0] / cltv_c.shape[0]
    churn_rate = 1 - repeat_rate

    # 8. Kâr Marjı (Profit Margin) (toplam harcamanın belirli bir oranı)
    cltv_c['profit_margin'] = cltv_c['total_price'] * profit

    # 9. Müşteri Değeri (Customer Value)
    cltv_c['customer_value'] = cltv_c['avg_order_value'] * cltv_c["purchase_frequency"]

    # 10. CLTV (Customer Lifetime Value) Hesaplama
    cltv_c['cltv'] = (cltv_c['customer_value'] / churn_rate) * cltv_c['profit_margin']

    # 11. CLTV değerine göre 4 segmente böler (A en iyi, D en düşük)
    cltv_c["segment"] = pd.qcut(cltv_c["cltv"], 4, labels=["D", "C", "B", "A"])

    return cltv_c
````

````python
df = df_.copy()

clv = create_cltv_c(df)
````
