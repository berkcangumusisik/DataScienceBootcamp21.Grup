## Müşteri Yaşam Boyu Değeri (CLTV) Tahmini Nedir?

Müşteri Yaşam Boyu Değeri (Customer Lifetime Value - CLTV), bir müşterinin bir işletmeyle olan ilişkisi boyunca sağlayacağı toplam finansal değerin tahminidir. Başka bir deyişle, bir müşterinin şirketten ortalama olarak kazandıracağı gelir miktarını temsil eder.

CLTV tahmini, müşteri kazanımı ve tutundurulması maliyetlerini optimize etmek, pazarlama stratejilerini geliştirmek ve müşteri segmentasyonu yapmak için önemli bir analiz aracıdır. Uzun vadeli kârlılığı ve sürdürülebilir büyümeyi desteklemek adına işletmeler CLTV tahminini kullanarak;

- En kârlı müşteri segmentlerini belirleyebilir,
- Hangi müşterilere daha fazla yatırım yapılacağını anlayabilir,
- Özelleştirilmiş pazarlama stratejileri geliştirebilir,
- Müşteri sadakati ve elde tutma stratejilerini efektif şekilde planlayabilirler.

Kısacası, CLTV tahmini sayesinde firmalar, müşteri odaklı kararlarını daha sağlıklı ve veri temelli verebilir ve rekabet avantajı elde edebilirler.

## CLTV (Customer Lifetime Value) Hesaplama Formülü

CLTV formülü temel olarak aşağıdaki gibidir:

**CLTV = (Customer Value / Churn Rate) x Profit Margin**

Buradaki ana kavramları açacak olursak:

- **Customer Value (Müşteri Değeri)**: Bir müşterinin yaratacağı toplam değeri gösterir ve şu şekilde hesaplanır:

    **Customer Value = Purchase Frequency x Average Order Value**

    Yani, bir müşterinin alışveriş yapma sıklığı ile ortalama sipariş tutarının çarpımıdır.

- **CLTV'nin Alternatif Bir Hesaplanışı (Olasılıksal Modellerde):**

    **CLTV = Expected Number of Transaction x Expected Average Profit**

    Burada müşteri başına beklenen toplam işlem sayısı ile beklenen ortalama kâr çarpılarak müşterinin firma için tahmini yaşam boyu değeri bulunabilir.

Bu formüller sayesinde, bir müşterinin şirket ile ilişkisi boyunca sağlayacağı toplam finansal değeri tahmin edebiliriz.

## BG/NBD Modeli Nedir?

BG/NBD (Beta Geometric/Negative Binomial Distribution) modeli, müşterilerin gelecekte gerçekleştirecekleri alışveriş sayısını tahmin etmeye yarayan olasılıksal bir modeldir. Model, müşteri davranışlarını analiz ederek kimin tekrar alışveriş yapacağını ve kimin şirketle ilişkisini sonlandıracağını (churn) tahmin etmeye çalışır.

BG/NBD modelinde iki temel süreç bulunur: **Transaction Process (Buy)** ve **Dropout Process (Till you die)**.

**Transaction Process (Buy):**

- Bir müşteri alive olduğu sürece, belirli bir zaman periyodunda müşteri tarafından gerçekleştirilecek işlem sayısı, transaction rate (işlem oranı) parametresiyle Poisson dağılımı kullanılarak modellenir.
- Her müşteri alive olduğu sürece kendi transaction rate'i etrafında rastgele satın alma yapmaya devam eder.
- Transaction rate'ler her bir müşteri için farklılık gösterebilir ve tüm kitle için gamma dağılımı [Gamma(r, a)] ile dağıldığı varsayılır.

**Dropout Process (Till you die):**

- Her müşterinin p olasılığı ile ilişkiyi kesme oranı (dropout probability) vardır.
- Müşteri, alışveriş yaptıktan sonra belirli bir dropout probability ile churn olur (müşteriliğini sonlandırır).
- Dropout rate'ler de müşteri bazında farklılık gösterebilir ve tüm kitle için beta dağılımı [Beta(a, b)] ile dağıldığı varsayılır.


Bu model; işlem geçmişi kullanılarak, abonelik sistemi olmayan ve müşterilerin rastgele zamanlarda alışveriş yaptığı iş kollarında, bir müşterinin gelecekte beklenen alışveriş sayısının tahmini için yaygın olarak kullanılır. BG/NBD modelinden elde edilen alışveriş sayısı tahmini, müşteri yaşam boyu değerinin (CLTV) hesaplanmasında temel yapı taşlarından biridir.

## Gamma-Gamma Modeli Nedir?

Gamma-Gamma modeli, müşterilerin ortalama işlem/alışveriş başı harcama miktarını (Average Monetary Value) tahmin etmek için kullanılan olasılıksal bir modeldir.

Bu model genellikle BG/NBD veya Pareto/NBD gibi modellerle birlikte kullanılır. BG/NBD, bir müşterinin gelecekte kaç kez alışveriş yapacağını öngörürken; Gamma-Gamma modeli ise, bu alışverişlerde müşteriden beklenen ortalama harcama tutarını tahmin eder.

### Temel Varsayımlar:
- Bir müşterinin işlem başı harcama tutarları, o müşteriye özel ve belirli bir ortalamaya (μ) sahiptir.
- Bu ortalama harcama tutarı (μ), müşteri popülasyonu içinde Gamma dağılımına uyar.
- Tüm müşteriler için harcama değerleri bağımsızdır.
- Bir müşterinin işlem başı harcama değerleri, zaman içinde sabittir ve işlemler arası harcama miktarları müşteriler arası farklılık gösterebilir.

### Gamma-Gamma Modelinin Avantajları:
- Satın alma sıklığı (BG/NBD ile) çok, ancak harcama tutarları değişken olan müşteri gruplarında daha doğru tahminler sunar.
- Sadece bağımsız ve tekrarlayan işlemler için uygundur, abonelik veya toplu ödemeler için ideal değildir.

### Matematiksel Olarak:
Gamma-Gamma modeli, iki parametreli bir Gamma dağılımı ve harcama davranışına ilişkin bir hiperparametre ile tahminleme yapar. Model, her müşteri için beklenen ortalama işlem değerini (expected average profit) verir:

$$ \text{Expected Average Profit} = \frac{p + x}{q + n} $$

Burada:
- \( p, q \): Gamma dağılımı hiperparametreleri,
- \( x \): Toplam harcama,
- \( n \): İşlem sayısı olarak kullanılabilir.

### Özetle:
Gamma-Gamma modeli, müşterilerin harcama davranışını daha iyi modelleyerek, CLTV (Customer Lifetime Value) hesaplamalarında tahminlerin doğruluğunu yüksek oranda artırır.

```python
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
from lifetimes import BetaGeoFitter
from lifetimes import GammaGammaFitter
from lifetimes.plotting import plot_period_transactions

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
pd.set_option('display.float_format', lambda x: '%.4f' % x)
from sklearn.preprocessing import MinMaxScaler
```

```python
def outlier_thresholds(dataframe, variable):
    """
    Bu fonksiyon, verilen dataframe ve değişken için aykırı (outlier) değer eşiklerini belirler.
    Kısaca fonksiyonun işleyişi:
    - 0.01 ve 0.99 çeyrek değerlerini (quantile) bulur.
    - İki çeyrek arasındaki farkı (interquantile range) hesaplar.
    - Alt ve üst sınırı, bu aralığa göre ayarlar.
    Böylece ilgili değişkende çok uçta kalan, veri dağılımının dışında olan değerleri tespit etmek için alt ve üst eşik değerleri döner.

    Args:
        dataframe (DataFrame): Üzerinde işlem yapılacak veri seti.
        variable (str): Aykırı değeri kontrol edilecek sütunun ismi.

    Returns:
        tuple: (alt_sınır, üst_sınır)
    """
    quartile1 = dataframe[variable].quantile(0.01)
    quartile3 = dataframe[variable].quantile(0.99)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
```

```python
def replace_with_thresholds(dataframe, variable):
    """
    Bu fonksiyon, verilen dataframe ve değişken için aykırı değerleri alt ve üst eşik değerleriyle değiştirir.
    Kısaca fonksiyonun işleyişi:
    - Verilen değişkenin aykırı değer eşiklerini (alt ve üst sınır) belirler.

    Args:
        dataframe (DataFrame): Üzerinde işlem yapılacak veri seti.
        variable (str): Aykırı değeri kontrol edilecek sütunun ismi.

    Returns:
        None
    """
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit


```

```python
df_ = pd.read_excel("datasets/online_retail_II.xlsx",
                    sheet_name="Year 2010-2011")
df = df_.copy()
df.describe().T
```

```python
df.head()
```

```python
# Veri Ön İşleme

df.dropna(inplace=True)
```

```python
df.describe().T
```

```python
df = df[~df["Invoice"].str.contains("C", na=False)] 
df
```

```python
df = df[df["Quantity"] > 0]
```

```python
df = df[df["Price"] > 0]
```

```python
df["Quantity"] = df["Quantity"].astype(float)
df["Price"] = df["Price"].astype(float)

replace_with_thresholds(df, "Quantity")
replace_with_thresholds(df, "Price")
df
```

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]
df.head()
```

```python
today_date = dt.datetime(2011, 12, 11)
```

```python
#########################
# Lifetime Veri Yapısının Hazırlanması
#########################

# recency: Son satın alma üzerinden geçen zaman. Haftalık. (kullanıcı özelinde)
# T: Müşterinin yaşı. Haftalık. (analiz tarihinden ne kadar süre önce ilk satın alma yapılmış)
# frequency: tekrar eden toplam satın alma sayısı (frequency>1)
# monetary: satın alma başına ortalama kazanç
cltv_df = df.groupby('Customer ID').agg(
    {'InvoiceDate': [lambda InvoiceDate: (InvoiceDate.max() - InvoiceDate.min()).days,
                     lambda InvoiceDate: (today_date - InvoiceDate.min()).days],
     'Invoice': lambda Invoice: Invoice.nunique(),
     'TotalPrice': lambda TotalPrice: TotalPrice.sum()})

cltv_df
```

```python
cltv_df.columns = ['recency', 'T', 'frequency', 'monetary']
cltv_df.head()
```

```python
cltv_df["monetary"] = cltv_df["monetary"] / cltv_df["frequency"]

cltv_df.describe().T
```

```python
cltv_df = cltv_df[(cltv_df['frequency'] > 1)]

cltv_df["recency"] = cltv_df["recency"] / 7

cltv_df["T"] = cltv_df["T"] / 7

cltv_df.describe().T
```

### 2. BG-NBD Modelinin Kurulması


```python
# BG-NBD (Beta-Geometric/Negative Binomial Distribution) modeli müşteri satın alma sıklığını tahmin etmeye yarayan bir yaşam boyu değeri (CLTV) tahminlemesinde kullanılan istatistiksel bir modeldir. 
# Bu aşamada model nesnesi oluşturulup penalizer_coef ile aşırı öğrenmenin (overfitting) önüne geçilmesi amaçlanır. 
# Ardından model, frequency (tekrar eden satın alma sayısı), recency (son alışveriş ile ilk alışveriş arasındaki süre) ve T (müşterinin şirkette geçirdiği süre) verileriyle eğitilir.

bgf = BetaGeoFitter(penalizer_coef=0.001)

bgf.fit(cltv_df['frequency'],
        cltv_df['recency'],
        cltv_df['T'])
```

```python
# 1 hafta içinde en çok satın alma beklediğimiz 10 müşteri kimdir?
# Aşağıdaki kod, BG-NBD modelimiz tarafından önümüzdeki 1 hafta içinde en fazla satın alma gerçekleştirmesi beklenen 10 müşteriyi gösterir.
# conditional_expected_number_of_purchases_up_to_time fonksiyonu, verilen zaman aralığında (burada 1 hafta), her müşteri için tahmini alışveriş sayısını hesaplar.
# Sonrasında bu tahminler azalan şekilde sıralanır ve en yüksek ilk 10 müşteri listelenir.

bgf.conditional_expected_number_of_purchases_up_to_time(
    1,
    cltv_df['frequency'],
    cltv_df['recency'],
    cltv_df['T']
)

```

```python
bgf.predict(1,
            cltv_df['frequency'],
            cltv_df['recency'],
            cltv_df['T']).sort_values(ascending=False).head(10)
```

```python
cltv_df["expected_purc_1_week"] = bgf.predict(1,
                                              cltv_df['frequency'],
                                              cltv_df['recency'],
                                              cltv_df['T'])

cltv_df["expected_purc_1_week"].sort_values(ascending=False).head(10)
```

```python
cltv_df["expected_purc_1_month"] = bgf.predict(4,
                                               cltv_df['frequency'],
                                               cltv_df['recency'],
                                               cltv_df['T'])
```

```python
bgf.predict(4,
            cltv_df['frequency'],
            cltv_df['recency'],
            cltv_df['T']).sum()
```

```python
# 3 Ayda Tüm Şirketin Beklenen Satış Sayısı Nedir?
# Aşağıdaki kod, BG-NBD modelinin öngördüğü şekilde, önümüzdeki 3 ay (12 hafta) boyunca tüm müşteriler için tahmini toplam satın alma sayısını hesaplar ve toplar.
# predict fonksiyonu, her bir müşteri için beklenen satın alma sayısını döndürürken, sum() fonksiyonu bunları toplayarak şirket genelinde toplamı elde etmemizi sağlar.

bgf.predict(4 * 3,
            cltv_df['frequency'],
            cltv_df['recency'],
            cltv_df['T']).sum()
```

```python
cltv_df["expected_purc_3_month"] = bgf.predict(4 * 3,
                                               cltv_df['frequency'],
                                               cltv_df['recency'],
                                               cltv_df['T'])

```

### 3. GAMMA-GAMMA Modelinin Kurulması


```python
ggf = GammaGammaFitter(penalizer_coef=0.01)

ggf.fit(cltv_df['frequency'], cltv_df['monetary'])
```

```python
# Bu kod, Gamma-Gamma modelini kullanarak her bir müşterinin beklenen ortalama kârını (expected average profit) hesaplar ve ilk 10 müşterinin sonucunu gösterir.
ggf.conditional_expected_average_profit(cltv_df['frequency'],
                                        cltv_df['monetary']).head(10)
                                   
```

```python
cltv_df["expected_average_profit"] = ggf.conditional_expected_average_profit(cltv_df['frequency'],
                                                                             cltv_df['monetary'])
cltv_df.sort_values("expected_average_profit", ascending=False).head(10)
```

### 4. BG-NBD ve GG modeli ile CLTV'nin hesaplanması.

```python
cltv = ggf.customer_lifetime_value(bgf,
                                   cltv_df['frequency'],
                                   cltv_df['recency'],
                                   cltv_df['T'],
                                   cltv_df['monetary'],
                                   time=4 * 3, # 4 hafta
                                   freq="W", # haftalık
                                   discount_rate=0.01)
cltv.head()

```

```python
cltv = cltv.reset_index()

cltv_final = cltv_df.merge(cltv, on="Customer ID", how="left")
cltv_final.sort_values(by="clv", ascending=False).head(10)
```

### 5. CLTV'ye Göre Segmentlerin Oluşturulması


```python
cltv_final["segment"] = pd.qcut(cltv_final["clv"], 4, labels=["D", "C", "B", "A"])

cltv_final.sort_values(by="clv", ascending=False).head(50)
```

### 6. Çalışmanın Fonksiyonlaştırılması


```python
def create_cltv_p(dataframe, month=3):
    """
    Bu fonksiyon, verilen bir satış verisi DataFrame üzerinde Müşteri Yaşam Boyu Değeri (Customer Lifetime Value, CLTV) tahmini yapar.
    
    Temel adımlar şunlardır:
    1. Veri Ön İşleme: Eksik veriler temizlenir, iadeler hariç tutulur, miktar ve fiyat negatif değerlerden arındırılır, aykırı değerler baskılanır, TotalPrice hesaplanır. 
    2. CLTV için temel değişkenlerin oluşturulması: Her bir müşterinin özet (recency, T, frequency, monetary) metrikleri hesaplanır ve haftaya çevrilir.
    3. BG/NBD Modeli kurulumu: Müşteri alışveriş sıklığını ve tekrar olasılığını tahmin eder. 
    4. Gamma-Gamma Modeli kurulumu: Kullanıcı başı ortalama kârlılığı tahmin eder.
    5. CLTV (Customer Lifetime Value) hesaplanması: BG-NBD ve Gamma-Gamma tahminlerinden parasal değer elde edilir. 
    6. Sonuçta her müşterinin segmenti (A, B, C, D) çıkartılır.
    Fonksiyonun çıktısı, müşteri seviyesinde tüm ana CLTV parametrelerini ve segmentleri içeren DataFrame'dir.

    Parametreler:
    dataframe (pd.DataFrame): Satış verilerini içeren dataframe.
    month (int): CLTV'nin tahmin edilmek istendiği toplam ay sayısı (ör: 3 = 3 aylık tahmin).

    Dönen Değer:
    pd.DataFrame: Müşteri bazında CLTV özetlerini ve segmentlerini içerir.
    """
    # 1. Veri Ön İşleme
    dataframe.dropna(inplace=True)  # Eksik değerleri çıkar
    dataframe = dataframe[~dataframe["Invoice"].str.contains("C", na=False)]  # İade faturalarını çıkar (Invoice içinde 'C' olanlar)
    dataframe = dataframe[dataframe["Quantity"] > 0]  # Negatif miktarları çıkar
    dataframe = dataframe[dataframe["Price"] > 0]     # Negatif fiyatları çıkar
    replace_with_thresholds(dataframe, "Quantity")    # Aykırı miktarları baskıla
    replace_with_thresholds(dataframe, "Price")       # Aykırı fiyatları baskıla
    dataframe["TotalPrice"] = dataframe["Quantity"] * dataframe["Price"]  # Satır bazında toplam harcama
    today_date = dt.datetime(2011, 12, 11)            # Son alışveriş tarihini kıyaslamak için sabit "bugün" tarihi

    # 2. Müşteri bazında temel CLTV değişkenleri (recency, T, frequency, monetary) hesaplanır
    cltv_df = dataframe.groupby('Customer ID').agg({
        'InvoiceDate': [lambda InvoiceDate: (InvoiceDate.max() - InvoiceDate.min()).days,
                        lambda InvoiceDate: (today_date - InvoiceDate.min()).days],
        'Invoice'    : lambda Invoice: Invoice.nunique(),
        'TotalPrice' : lambda TotalPrice: TotalPrice.sum()
    })
    cltv_df.columns = cltv_df.columns.droplevel(0)
    cltv_df.columns = ['recency', 'T', 'frequency', 'monetary']
    cltv_df["monetary"] = cltv_df["monetary"] / cltv_df["frequency"]  # İşlem başına ort. kazanç
    cltv_df = cltv_df[(cltv_df['frequency'] > 1)]  # 1'den fazla alışveriş yapanlar filtrelendi
    cltv_df["recency"] = cltv_df["recency"] / 7  # Gün hesabı haftaya çevrildi
    cltv_df["T"] = cltv_df["T"] / 7

    # 3. BG-NBD Modelinin Kurulması: Tekrar alışveriş sayıları ve olasılıkları tahmin edilir
    bgf = BetaGeoFitter(penalizer_coef=0.001)
    bgf.fit(cltv_df['frequency'],
            cltv_df['recency'],
            cltv_df['T'])

    # Haftalık, aylık ve üç aylık beklenen satın alma sayıları hesaplanır
    cltv_df["expected_purc_1_week"] = bgf.predict(1,
                                                  cltv_df['frequency'],
                                                  cltv_df['recency'],
                                                  cltv_df['T'])
    cltv_df["expected_purc_1_month"] = bgf.predict(4,
                                                   cltv_df['frequency'],
                                                   cltv_df['recency'],
                                                   cltv_df['T'])
    cltv_df["expected_purc_3_month"] = bgf.predict(12,
                                                   cltv_df['frequency'],
                                                   cltv_df['recency'],
                                                   cltv_df['T'])

    # 4. GAMMA-GAMMA Modelinin Kurulması: Ortalama işlem başına karlılık hesaplanır
    ggf = GammaGammaFitter(penalizer_coef=0.01)
    ggf.fit(cltv_df['frequency'], cltv_df['monetary'])
    cltv_df["expected_average_profit"] = ggf.conditional_expected_average_profit(cltv_df['frequency'],
                                                                                cltv_df['monetary'])

    # 5. BG-NBD ve GG modelleriyle CLTV hesaplanır
    cltv = ggf.customer_lifetime_value(bgf,
                                       cltv_df['frequency'],
                                       cltv_df['recency'],
                                       cltv_df['T'],
                                       cltv_df['monetary'],
                                       time=month,            # Tahmin yapılacak toplam ay
                                       freq="W",              # Haftalık olarak
                                       discount_rate=0.01)    # İskonto oranı
    
    cltv = cltv.reset_index()  # CLTV skorunda müşteri ID'si index'te olduğu için sütuna al
    cltv_final = cltv_df.merge(cltv, on="Customer ID", how="left")  # Tüm verileri birleştir

    # 6. CLTV skorlarına göre müşteri segmentasyonu yapılır
    cltv_final["segment"] = pd.qcut(cltv_final["clv"], 4, labels=["D", "C", "B", "A"])

    return cltv_final
```
